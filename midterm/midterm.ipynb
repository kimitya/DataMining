```python
X_data= data.drop('Weather Type', axis=1)
y_data = data['Weather Type']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2)
print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
#22B030380
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(X_train, y_train)
#22B030380
```

```python
feature_importances = rfc.feature_importances_
importance_df = pd.DataFrame({
    'Feature': X_data.columns,
    'Importance': feature_importances
})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
#22B030380
```

```python
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Weather Feature Importance from Random Forest')
plt.gca().invert_yaxis()
plt.show()
#22B030380
```

```python
lowest_importance_features = importance_df.tail(4)['Feature'].values
X_data_reduced = X_data.drop(columns=lowest_importance_features)
X_data_reduced.head()
#22B030380
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data=pd.DataFrame(scaler.fit_transform(data))
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_reduced, y_data, test_size=0.2)
#22B030380
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

classifiers = {
    'Decision Tree': DecisionTreeClassifier(),
    #default solver didn't work
    'Logistic Regression': LogisticRegression(max_iter=1000, solver='liblinear')
}
#22B030380
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

advanced_classifiers = {
    'SVM': SVC(kernel='linear'),
    'Random Forest': RandomForestClassifier(),
    'Ada Boost': AdaBoostClassifier()
}
#22B030380
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

model_names = []
accuracies = []
confusion_matrices = {}
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    model_names.append(name)
    accuracies.append(accuracy)
    cm = confusion_matrix(y_test, y_pred)
    confusion_matrices[name] = cm
    print(f"{name} Accuracy: {accuracy:.4f}")
#22B030380
```

```python
for name, clf in advanced_classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    model_names.append(name)
    accuracies.append(accuracy)
    cm = confusion_matrix(y_test, y_pred)
    confusion_matrices[name] = cm
    print(f"{name} Accuracy: {accuracy:.4f}")
#22B030380
```

```python
plt.figure(figsize=(10, 6))
plt.plot(model_names, accuracies, marker='o', linestyle='-', color='blue')
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])  
plt.grid(True)
plt.tight_layout()       
plt.show()
#22B030380
```

```python
for name, cm in confusion_matrices.items():
    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix: {name}")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```
