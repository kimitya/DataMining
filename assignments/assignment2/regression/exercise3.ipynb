Exercise 3: Decision Tree Regression
Objective: Build a decision tree regression model and visualize it.
Load the Boston Housing dataset from scikit-learn.
Split the dataset into training and testing sets.
Train a decision tree regressor on the training set.
Evaluate the model's performance using mean squared error (MSE).
Visualize the decision tree.

```python
from sklearn import datasets
df=datasets.load_iris()
df
```

```python
import pandas as pd
feature_names=df.feature_names
X_data=pd.DataFrame(df.data, columns=feature_names)
y_data=df.target
X_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = dtr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20, 10))
tree.plot_tree(dtr, feature_names=df.feature_names, filled=True)
plt.title("Decision Tree Regression Visualization")
plt.show()
```
